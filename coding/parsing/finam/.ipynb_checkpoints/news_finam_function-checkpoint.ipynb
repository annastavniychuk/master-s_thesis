### Библиотеки

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd 
from time import sleep
```

---

### Функция-парсер

```python
def news_parser(company, start_date, end_date): # вводить всё строками!
    template_url = 'https://www.finam.ru/profile/moex-akcii/'
    page = '1'
    url = template_url + company + '/news/?start-date=' + start_date + '&end-date=' + end_date + '&page=1'
    base_url = template_url + company + '/news/' # чтобы запихнуть в requests

    resp = requests.get(base_url, 
                    params={'start-date': start_date,
                            'end-date': end_date,
                            'page': page})
    soup = bs(resp.text)
    pages = soup.find('div', {'class': 'pager'}).find_all('li')[-2].text.split('...')[-1]
    
    print('Total pages:', pages, end='\n')
    print('Total news:', int(pages)*20, end='\n')
    
    news = []
    for i in range(1,int(pages)+1):
        print('Start page', i, end='\r')
        resp = requests.get(base_url, 
                            params={'start-date': start_date,
                                    'end-date': end_date,
                                    'page': i})
        #print('status', resp.status_code)
        soup = bs(resp.text)
        news_set = soup.find('tbody', {'class': 'news'}).find_all('tr')
        for item in news_set:
            date = item.find_all('td')[0].text
            preview = item.find_all('td')[1].text
            url = 'https://www.finam.ru' + item.find_all('td')[1].find('a').attrs['href']
            news.append({
                'preview_date': date,
                'preview_text': preview,
                'url': url,
                'company': company}) 
        sleep(1)
    
    print('', end='\n')
    
    for item in news:
        print('Start news', news.index(item), end='\r')
        resp = requests.get(item['url'])
        soup = bs(resp.text)
        article_time = soup.find('div', {'class': 'sm lightgrey mb05 mt15'}).text[11:-18]
        article_date = soup.find('div', {'class': 'sm lightgrey mb05 mt15'}).text[:-24]
        text_parts = soup.find('div', {'class': 'handmade mid f-newsitem-text'}).find_all('p')
        article = []
        for part in text_parts:
            article.append(part.text) 
        article = ' '.join(article)
        item.update({'article_time': article_time,
                     'article_date': article_date,
                     'article_text':article})
        #sleep(1)
    return news
```

### Фильтрация данных

```python
news = pd.read_csv('news.csv')
news.head()
```

```python
#news['FAS'] = news['article_text'].str.contains('антимонопол')
#news['FAS'] |= news['article_text'].str.contains('ФАС')
```

```python
news['FAS'] = news['article_text'].str.contains('федеральн.*? антимонопольн.*? служб', case=False)
news['FAS'] |= news['article_text'].str.contains('ФАС', case=False)
```

```python
news['event_predupr'] = news['article_text'].str.contains('предупр.*?', case=False)

news['event_predost'] = news['article_text'].str.contains('предостер.*?', case=False)

news['event_delo'] = news['article_text'].str.contains('возбу.*?', case=False)

news['event_resh'] = news['article_text'].str.contains('вынес.*? реш', case=False)
news['event_resh'] |= news['article_text'].str.contains('вынес.*?', case=False)

news['event_predp'] = news['article_text'].str.contains('предпис.*?', case=False)
```

```python
pd.DataFrame(news).to_csv('news_dummy_filter.csv')
pd.DataFrame(news).to_excel('news_dummy_filter.xlsx')
```

```python
filtered_news = news.loc[(news['FAS']==True) & ((news['event_predupr']==True)|(news['event_predost']==True)|(news['event_delo']==True)|(news['event_resh']==True)|(news['event_predp']==True))]
filtered_news.head()
```

```python
pd.DataFrame(filtered_news).to_csv('news_filtered.csv')
pd.DataFrame(filtered_news).to_excel('news_filtered.xlsx')
```
