```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd 
```

```python
resp = requests.get('https://www.finam.ru/profile/moex-akcii/rosneft/news/?start-date=2016-01-01&end-date=2016-12-31&page=1')
soup = bs(resp.text)
```

```python
news_set = soup.find('tbody', {'class': 'news'}).find_all('tr')
```

```python
news = []

for item in news_set:
    date = item.find_all('td')[0].text
    preview = item.find_all('td')[1].text
    url = 'https://www.finam.ru' + item.find_all('td')[1].find('a').attrs['href']
    news.append({
        'preview_date': date,
        'preview': preview,
        'url': url})
```

```python
pd.DataFrame(news)
```

```python
for item in news:
    resp = requests.get(item['url'])
    soup = bs(resp.text)
    calendar = soup.find('div', {'class': 'sm lightgrey mb05 mt15'}).text
    article_time = soup.find('div', {'class': 'sm lightgrey mb05 mt15'}).text[:-18].split()[1]
    article_date = soup.find('div', {'class': 'sm lightgrey mb05 mt15'}).text[:-18].split()[0]
    item.update({'article_time': article_time,
                 'article_date': article_date})
```

```python
pd.DataFrame(news)
```

```python
for item in news:
    resp = requests.get(item['url'])
    soup = bs(resp.text)
    text_parts = soup.find('div', {'class': 'handmade mid f-newsitem-text'}).find_all('p')
    article = []
    for part in text_parts:
        article += part.text
    item.update({'article':article})
```
